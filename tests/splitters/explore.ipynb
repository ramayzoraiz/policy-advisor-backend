```python
from pathlib import Path
def get_dir_above_target(p: Path, target_dir='tests'):
    for parent in p.parents:
        if parent.name == target_dir:
            return parent.parent
    # If p itself is target_dir
    if p.name == target_dir:
        return p.parent
    raise ValueError(f"'{target_dir}' not found in path")
```

```python
cwd = Path.cwd()
ywd=get_dir_above_target(cwd, target_dir='tests')
print(f"cwd before : {cwd}")
%cd {ywd}
print(f"ywd after : {ywd}")
```
